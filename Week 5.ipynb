```python
df.fillna({'Sleep_Hours': 0, 'Previous_Scores': 0, 'Tutoring_Sessions': 0,'Teacher_Quality': 'Unknown','Physical_Activity': 0, 'Parental_Education_Level': 'Not_Specified', 'Distance_from_Home': 'Unknown', 'Gender': 'Not_Specified'}, inplace=True)
```

```python
df.isna().sum()
```

```python
df2=df.copy()
```

```python
Q1=df2['Exam_Score'].quantile(0.25)
Q2=df2['Exam_Score'].quantile(0.75)
IQR=Q2-Q1

#define lower and upper limit
lower_limit=Q1-1.5*IQR
upper_limit=Q2+1.5*IQR

#Detect Outliers
outliers=df2[(df2['Exam_Score']<lower_limit) | (df2['Exam_Score']>upper_limit)]
print(outliers)
```
